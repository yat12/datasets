```python

df['relationship'] = df['relationship'].map({'Not-in-family': 0, 'Wife': 1, 'Other-relative': 2, 'Unmarried': 3,'Husband': 4,'Own-child': 5}).astype(int)



```

```python
df
```

```python
df.drop(['occupation'], axis=1, inplace=True)
```

```python
df
```

```python

df_x = pd.DataFrame(np.c_[df['relationship'], df['education'], df['race'],df['gender'],df['marital'],df['workclass']], columns = ['relationship','education','race','gender','marital','workclass'])

df_y = pd.DataFrame(df.income)
```

```python
reg = LogisticRegression()

x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.33, random_state=42)
```

```python
reg.fit(x_train, y_train)

y_pred = reg.predict(x_test)
```

```python
reg.predict([[1,7,3,7,0,2]])
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
